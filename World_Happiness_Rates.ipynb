```python
# Determining the independent features and dependent/predicted feature
X = df[['GDP per capita',
       'Social support',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Healthy life expectancy']]
y = df['Score']
```

```python
#Spliting the dataset 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=32)
```

```python
#Fitting the linear regression model 
from sklearn.linear_model import LinearRegression
```

```python
lm_df=LinearRegression()
```

```python
lm_df.fit(X_train, y_train)
```

```python
#Prediction
predict=lm_df.predict(X_test)
```

```python
plt.scatter(y_test,predict)
plt.xlabel('Y Test')
plt.ylabel('Predicted y')
```

```python
sns.distplot(predict-y_test)
```

```python
#Evaluating the model

from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(predict,y_test))
print('MSE:', metrics.mean_squared_error(predict,y_test))
print('RMSE:',np.sqrt(metrics.mean_squared_error(predict,y_test)))
```

```python
from sklearn import metrics
print ('Accuracy_Rate: ', metrics.r2_score(predict,y_test))
```
